```python
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(3,9,13)
#x= np.array([2,3,4,5,6])
x
```

```python
l=[]
for n in x:
    temp = (np.random.choice(range(1,3)))
    l.append(n+temp)
y= np.array(l)
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
sumX=np.sum(x)#sum of x
sumy=np.sum(y)#sum of y
x2sum=np.sum(x*x)
sumXY=np.sum(x*y)
sumX2=np.sum(x)**2# square of whole sum of x
#for finding the slope we have a formula
b=(len(x)*sumXY-sumX*sumy)/(len(x)*x2sum - sumX2)
b
```

```python
#h3n3c for the interceptor
a= (sumy - (b*sumX))/len(x)
a
```

```python
yp = a+b*x
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp,color='red')
plt.legend(["Orignal Y values","Predicted line"])
```
